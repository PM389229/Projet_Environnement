```python
#On replace les NA par des 0 , les valeurs non numeriques par N.A pour que ce soit plus facile a traiter
data['Natural gas production (bcm)'] = pd.to_numeric(data['Natural gas production (bcm)'], errors='coerce').fillna(0)
data['Coal and lignite domestic consumption (Mt)'] = pd.to_numeric(data['Coal and lignite domestic consumption (Mt)'], errors='coerce').fillna(0)
data['Share of wind and solar in electricity production (%)'] = pd.to_numeric(data['Share of wind and solar in electricity production (%)'], errors='coerce').fillna(0)
data['Coal and lignite production (Mt)'] = pd.to_numeric(data['Coal and lignite production (Mt)'], errors='coerce').fillna(0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(20,5), gridspec_kw={'width_ratios': [1, 1, 1]})
fig.subplots_adjust(wspace=0.4)

axes[0].set_title('CO2 emissions from fuel combustion (MtCO2) distribution')
sns.boxplot(data=data, y='CO2 emissions from fuel combustion (MtCO2)', ax=axes[0]).set_ylabel('')

sns.histplot(data=data, x='CO2 emissions from fuel combustion (MtCO2)', ax=axes[1]).set_xlabel('')

grouped_data = data.groupby('country')['CO2 emissions from fuel combustion (MtCO2)'].mean().reset_index()
sorted_data = grouped_data.nlargest(10, 'CO2 emissions from fuel combustion (MtCO2)').sort_values('CO2 emissions from fuel combustion (MtCO2)', ascending=False)

sns.barplot(data=sorted_data, x='CO2 emissions from fuel combustion (MtCO2)', y='country', ax=axes[2], errorbar=None, order=sorted_data['country']).set_ylabel('')
axes[1].set_ylabel("")
axes[2].set_xlabel("")

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, y='CO2 emissions from fuel combustion (MtCO2)', x = 'Year', hue = 'Region', estimator = 'mean', markers = False, errorbar=None)\
.set_title('CO2 emissions from fuel combustion (MtCO2)')
plt.xlabel("")
plt.ylabel("")
plt.show()

#On affiche ici les emissions de CO2 par regions du globe
```
